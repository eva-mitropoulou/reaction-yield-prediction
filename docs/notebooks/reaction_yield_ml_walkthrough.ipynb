## 1. Project Goal

Demonstrate reaction cleaning, categorical component featurization, leakage-aware validation, model benchmarking, uncertainty/error diagnostics, active-learning simulation, and existing-record ranking over public HTE records.

```python
from pathlib import Path
import json
import pandas as pd

ROOT = Path('..').resolve()
def load_json(path):
    with open(ROOT / path, 'r', encoding='utf-8') as handle:
        return json.load(handle)
```

```python
clean = pd.read_csv(ROOT / 'data/processed/clean_reactions.csv')
print('shape:', clean.shape)
print('columns:', clean.columns.tolist())
print('component_columns:', [c for c in clean.columns if c.startswith('component_')])
```

## 3. Aggregate Counts Only

```python
summary = {
    'row_count': int(clean.shape[0]),
    'component_cardinalities': {c: int(clean[c].nunique()) for c in clean.columns if c.startswith('component_')},
    'yield_min': float(clean['yield_percent'].min()),
    'yield_max': float(clean['yield_percent'].max()),
}
summary
```
